```python
from googletrans import Translator
import ipywidgets as widgets
from IPython.display import display, HTML
```

```python
translator=Translator()
```

```python
def translate_text(text,target_lang):
  translated=translator.translate(text,dest=target_lang)
  return translated.text
```

```python
text_input = widgets.Textarea(
    placeholder="Enter text to translate...",
    layout=widgets.Layout(
        width="500px",
        height="200px",
        border='2px solid #ccc',
        border_radius='8px',
        padding='10px'
    )
)

lang_dropdown = widgets.Dropdown(
    options=[
        ("Arabic", "ar"),
        ("Bengali", "bn"),
        ("Chinese (Simplified)", "zh-cn"),
        ("Chinese (Traditional)", "zh-tw"),
        ("Czech", "cs"),
        ("Danish", "da"),
        ("Dutch", "nl"),
        ("English", "en"),
        ("Finnish", "fi"),
        ("French", "fr"),
        ("German", "de"),
        ("Greek", "el"),
        ("Gujarati", "gu"),
        ("Hebrew", "he"),
        ("Hindi", "hi"),
        ("Hungarian", "hu"),
        ("Indonesian", "id"),
        ("Italian", "it"),
        ("Japanese", "ja"),
        ("Kannada", "kn"),
        ("Korean", "ko"),
        ("Malayalam", "ml"),
        ("Marathi", "mr"),
        ("Nepali", "ne"),
        ("Norwegian", "no"),
        ("Polish", "pl"),
        ("Portuguese", "pt"),
        ("Punjabi", "pa"),
        ("Romanian", "ro"),
        ("Russian", "ru"),
        ("Sinhala", "si"),
        ("Spanish", "es"),
        ("Swedish", "sv"),
        ("Tamil", "ta"),
        ("Telugu", "te"),
        ("Thai", "th"),
        ("Turkish", "tr"),
        ("Ukrainian", "uk"),
        ("Urdu", "ur"),
        ("Vietnamese", "vi")
    ],
    value="es",
    description="Translate to:",
    layout=widgets.Layout(width="300px")
)
```

```python
translate_button=widgets.Button(
    description="Translate",
    button_style='primary',
    tooltip='Click to translate',
    icon='language',
    layout=widgets.Layout(width='150px', margin='0 0 0 10px')
)
output_box=widgets.Output(
    layout=widgets.Layout(
        border='1px solid #ddd',
        padding='15px',
        margin='10px 0 0 0',
        min_height='100px',
        background_color='#f9f9f9',
        border_radius='8px'
    )
)
```

```python
def on_translate_button_clicked(b):
     output_box.clear_output()
     with output_box:
       display(HTML("""
       <style>
         @keyframes pulse {
           0% { transform: scale(1); opacity: 1; }
           50% { transform: scale(1.05); opacity: 0.7; }
           100% { transform: scale(1); opacity: 1; }
         }
         .pulsing-text {
           animation: pulse 1.5s infinite;
         }
       </style>
       <p class='pulsing-text' style='font-size: 18px; color: #666; font-weight: normal; background-color: #f0f0f0; padding: 10px; border-radius: 5px;'>Translating...</p>
       """))
     translated_text = translate_text(text_input.value,lang_dropdown.value)
     output_box.clear_output()
     with output_box:
       display(HTML(f"<p style='font-size: 18px; color: #333; font-weight: bold; background-color: #f0f0f0; padding: 10px; border-radius: 5px;'>Translated: {translated_text}</p>"))
```

```python
translate_button.on_click(on_translate_button_clicked)
```

```python

app_layout = widgets.VBox([
    widgets.HTML("<h2>Enter text below:</h2>"),
    text_input,
    widgets.HBox([lang_dropdown, translate_button], layout=widgets.Layout(justify_content='flex-start', margin='10px 0')), # Align items to the start
    widgets.HTML("<h2>Translated Text:</h2>"),
    output_box
],
layout=widgets.Layout(
    border='2px solid #007bff',
    padding='20px',
    margin='20px auto',
    border_radius='12px',
    background_color='#e0f2f7',
    width='fit-content',
    align_items='stretch'
))

display(app_layout)
```
